# Lab2. Design of Logic Gates using Perceptron and Keras

### Part-I: Design OR gate using the concept of Perceptron

```python
import numpy as np #importing necessary modules
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def logic_gate(w1, w2, b): #defining logic gate function and test function
    return lambda x1, x2: sigmoid(w1 * x1 + w2 * x2 + b)
def test(gate):
    for a, b in (0, 0), (0, 1), (1, 0), (1, 1):
        print("{}, {}: {}".format(a, b, np.round(gate(a,b))))
```

```python
or_gate = logic_gate(20, 20, -10)
```

```python
test(or_gate)
```

### Part-II: Implement the operations of AND, NOR and NAND gates

```python
#step 1 AND gate
and_gate = logic_gate(1,1,-1)
test(and_gate)
```

```python
#step 2 NAND gate
nand_gate=logic_gate(-1,-1,1.5)
test(nand_gate)
```

```python
#Step 3 NOR gate
nor_gate= logic_gate(-10,-10,5)
test(nor_gate)
```

### Part-III: Limitations of single neuron for XOR operation

```python
#Yes still we can build it by combining two logic gates and giving the output to the another logic gate.
def or_gate(x1, x2):
    return logic_gate(20, 20, -10)(x1, x2)
def nand_gate(x1, x2):
    return logic_gate(-20, -20, 30)(x1, x2)
def and_gate(x1, x2):
    return logic_gate(20, 20, -30)(x1, x2)
def xor_gate(x1, x2):
    return and_gate(or_gate(x1, x2), nand_gate(x1, x2))
for a, b in [(0, 0), (0, 1), (1, 0), (1, 1)]:
    print("{}, {}: {}".format(a, b, np.round(xor_gate(a, b))))
```

### Part-IV: Logic Gates using Keras library

```python
from keras.models import Sequential #building DNN model for And gate
from keras.layers import Dense
input_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
output_data = np.array([[0], [0], [0], [1]])

model = Sequential() #creating the instance of the model
model.add(Dense(16, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['binary_accuracy'])
model.fit(input_data, output_data, epochs=100) #fitting the model

predictions = model.predict(input_data) #model prediction
print(predictions)
```
