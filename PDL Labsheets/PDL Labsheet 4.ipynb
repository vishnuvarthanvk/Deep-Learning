# PDL Lab4.: Image corpus creation and binary classification using DNN

```python
# Importing necessary modules
import os 
import cv2
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras.optimizers import Adam
from sklearn.model_selection import train_test_split
```

```python
#Path to folder
rose_f = "C:\\Users\\Admin\\Downloads\\white rose"
cotton_f = 'C:\\Users\\Admin\\Downloads\\cotton'
```

```python
#Normalizing Images
images = []
labels = []

for filename in os.listdir(rose_f):
    if not filename.endswith(('.jpg', '.jpeg', '.png')):
        continue
    
    image_path = os.path.join(rose_f, filename)
    image = cv2.imread(image_path)
    image = cv2.resize(image, (500, 500))
    image = image / 255.0
    
    images.append(image)
    labels.append(0)  
```

```python
for filename in os.listdir(cotton_f):
    if not filename.endswith(('.jpg', '.jpeg', '.png')):
        continue
    
    image_path = os.path.join(cotton_f, filename)
    image = cv2.imread(image_path)
    image = cv2.resize(image, (500, 500))
    image = image / 255.0
    
    images.append(image)
    labels.append(1)  
```

```python
# Convert the lists to numpy arrays
images = np.array(images)
labels = np.array(labels)
```

```python
X_train,X_test,y_train,y_test=train_test_split(images,labels,test_size=0.25)
```

```python
# Define the model
model = Sequential()
model.add(Flatten(input_shape=(500, 500, 3)))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32)
```

```python
# Evaluate the model on test set
loss, accuracy = model.evaluate(X_test, y_test)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```

```python
# Predict class labels for test images
test_predictions = model.predict(X_test)
```

```python
# Convert probabilities to class labels (0 or 1)
test_predictions = np.round(test_predictions).flatten()
```

```python
# Print the predicted labels and the actual labels
print("Predicted Labels:", test_predictions)
print("Actual Labels:", y_test)
```

```python
# Save the image corpus and labels
np.save('image_corpus.npy', images)
np.save('labels.npy', labels)
```
